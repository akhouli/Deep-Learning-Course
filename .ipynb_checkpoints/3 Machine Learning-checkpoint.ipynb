```python
def line(x, w=0, b=0):
    return x * w + b
```

```python
x = np.linspace(55, 80, 100)
```

```python
yhat = line(x, w=0, b=0)
```

```python
yhat
```

```python
df.plot(kind='scatter',
        x='Height',
        y='Weight',
        title='Weight and Height in adults')
plt.plot(x, yhat, color='red', linewidth=3)
```

### Cost Function

```python
def mean_squared_error(y_true, y_pred):
    s = (y_true - y_pred)**2
    return s.mean()
```

```python
X = df[['Height']].values
y_true = df['Weight'].values
```

```python
y_true
```

```python
y_pred = line(X)
```

```python
y_pred
```

```python
mean_squared_error(y_true, y_pred.ravel())
```

Chane the values of the parameters b and w in the line above and plot it again to see how the plot and the cost  change.

```python
plt.figure(figsize=(10, 5))

# we are going to draw 2 plots in the same figure
# first plot, data and a few lines
ax1 = plt.subplot(121)
df.plot(kind='scatter',
        x='Height',
        y='Weight',
        title='Weight and Height in adults', ax=ax1)

# let's explore the cost function for a few values of b between -100 and +150
bbs = np.array([-100, -50, 0, 50, 100, 150])
mses = []  # we will append the values of the cost here, for each line
for b in bbs:
    y_pred = line(X, w=2, b=b)
    mse = mean_squared_error(y_true, y_pred)
    mses.append(mse)
    plt.plot(X, y_pred)

# second plot: Cost function
ax2 = plt.subplot(122)
plt.plot(bbs, mses, 'o-')
plt.title('Cost as a function of b')
plt.xlabel('b')
```

## Linear Regression with Keras

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam, SGD
```

```python
model = Sequential()
```

```python
model.add(Dense(1, input_shape=(1,)))
```

```python
model.summary()
```

```python
model.compile(Adam(lr=0.8), 'mean_squared_error')
```

```python
model.fit(X, y_true, epochs=40)
```

```python
y_pred = model.predict(X)
```

```python
df.plot(kind='scatter',
        x='Height',
        y='Weight',
        title='Weight and Height in adults')
plt.plot(X, y_pred, color='red')
```

```python
W, B = model.get_weights()
```

```python
W
```

```python
B
```

## Evaluating Model Performance

```python
from sklearn.metrics import r2_score
```

```python
print("The R2 score is {:0.3f}".format(r2_score(y_true, y_pred)))
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_true,
                                                    test_size=0.1)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
W[0, 0] = 0.0
B[0] = 0.0
model.set_weights((W, B))
```

```python
model.fit(X_train, y_train, epochs=50, verbose=0)
```

```python
y_train_pred = model.predict(X_train).ravel()
y_test_pred = model.predict(X_test).ravel()
```

```python
from sklearn.metrics import mean_squared_error as mse
```

```python
print("The Mean Squared Error on the Train set is:\t{:0.1f}".format(mse(y_train, y_train_pred)))
print("The Mean Squared Error on the Test set is:\t{:0.1f}".format(mse(y_test, y_test_pred)))
```

```python
print("The R2 score on the Train set is:\t{:0.3f}".format(r2_score(y_train, y_train_pred)))
print("The R2 score on the Test set is:\t{:0.3f}".format(r2_score(y_test, y_test_pred)))
```

# Classification

```python
df = pd.read_csv('../data/user_visit_duration.csv')
```

```python
df.head()
```

```python
df.plot(kind='scatter', x='Time (min)', y='Buy')
```

```python
model = Sequential()
model.add(Dense(1, input_shape=(1,), activation='sigmoid'))
```

```python
model.compile(SGD(lr=0.5), 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
X = df[['Time (min)']].values
y = df['Buy'].values

model.fit(X, y, epochs=25)
```

```python
ax = df.plot(kind='scatter', x='Time (min)', y='Buy',
             title='Purchase behavior VS time spent on site')

temp = np.linspace(0, 4)
ax.plot(temp, model.predict(temp), color='orange')
plt.legend(['model', 'data'])
```

```python
temp_class = model.predict(temp) > 0.5
```

```python
ax = df.plot(kind='scatter', x='Time (min)', y='Buy',
             title='Purchase behavior VS time spent on site')

temp = np.linspace(0, 4)
ax.plot(temp, temp_class, color='orange')
plt.legend(['model', 'data'])
```

```python
y_pred = model.predict(X)
y_class_pred = y_pred > 0.5
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("The accuracy score is {:0.3f}".format(accuracy_score(y, y_class_pred)))
```

### Train/Test split


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
params = model.get_weights()
params = [np.zeros(w.shape) for w in params]
model.set_weights(params)
```

```python
print("The accuracy score is {:0.3f}".format(accuracy_score(y, model.predict(X) > 0.5)))
```

```python
model.fit(X_train, y_train, epochs=25, verbose=0)
```

```python
print("The train accuracy score is {:0.3f}".format(accuracy_score(y_train, model.predict(X_train) > 0.5)))
print("The test accuracy score is {:0.3f}".format(accuracy_score(y_test, model.predict(X_test) > 0.5)))
```

## Cross Validation

```python
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
def build_logistic_regression_model():
    model = Sequential()
    model.add(Dense(1, input_shape=(1,), activation='sigmoid'))
    model.compile(SGD(lr=0.5),
                  'binary_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn=build_logistic_regression_model,
                        epochs=25,
                        verbose=0)
```

```python
from sklearn.model_selection import cross_val_score, KFold
```

```python
cv = KFold(3, shuffle=True)
```

```python
scores = cross_val_score(model, X, y, cv=cv)
```

```python
scores
```

```python
print("The cross validation accuracy is {:0.4f}  {:0.4f}".format(scores.mean(), scores.std()))
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y, y_class_pred)
```

```python
def pretty_confusion_matrix(y_true, y_pred, labels=["False", "True"]):
    cm = confusion_matrix(y_true, y_pred)
    pred_labels = ['Predicted '+ l for l in labels]
    df = pd.DataFrame(cm, index=labels, columns=pred_labels)
    return df
```

```python
pretty_confusion_matrix(y, y_class_pred, ['Not Buy', 'Buy'])
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
print("Precision:\t{:0.3f}".format(precision_score(y, y_class_pred)))
print("Recall:  \t{:0.3f}".format(recall_score(y, y_class_pred)))
print("F1 Score:\t{:0.3f}".format(f1_score(y, y_class_pred)))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y, y_class_pred))
```

```python
#Reading the file
df= pd.read_csv('../data/HR_comma_sep.csv')
df.head()
#we can see that all values are numeric except sales and salary
```

```python
# Establish a benchmark --> i.e what is the easiest thing
# I could do, the quickest model i could build
#left is the column we will use to predict whether some 1 left
df.left.value_counts() / len(df)
#24% of the people have left
# in other words, predicting that everyone has stayed
# yeilds an accuracy of 76%
# i.e our benchmark is we have to be better than 76%
```

```python
#Check if any feature needs rescaling,
# = monthly avg hours
# You should plot a histogram of the feature to decide which
# rescaling method
df['average_montly_hours'].plot(kind='hist')
```

```python
# after plotting, we can see that they're between 100 and 300
# so we can rescale them just by /100
df['average_montly_hours_100'] = df['average_montly_hours']/100.0
```

```python
df['average_montly_hours_100'].plot(kind='hist')
#replot, and see that they're between 1 and 3
```

```python
df['time_spend_company'].plot(kind='hist')
#GOOD, no need to be rescaled
```

```python
#Convert categorical features into binary dummy columns
# We will then have to combine them with 
# The numerical features using 'pd.concat'
df_dummies= pd.get_dummies(df[['sales','salary']])
```

```python
df_dummies.head()
```

```python
#our original columns
df.columns
```

```python
X= pd.concat([df[['satisfaction_level', 'last_evaluation', 
                  'number_project', 'time_spend_company', 'Work_accident',
                   'promotion_last_5years','average_montly_hours_100']],
              df_dummies],axis=1).values
y= df['left'].values
```

```python
X.shape
#we Can see that we have around 15000 points and 20 features
# and one variable to pred
```

```python
#Now, we do the usual train/test split with 20% test size
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
#then we build the model | predicting Binary (classification)
# so logistic regression problem
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
model = Sequential()
model.add(Dense(1, input_shape=(20,), activation='sigmoid'))
model.compile(SGD(lr=0.5), 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
#21 parameters --> 20 features and 1 bias
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
y_test_pred = model.predict_classes(X_test)
```

```python
# we can do a confusion matrix between our test predictions 
# and our true variables(test labels)
from sklearn.metrics import confusion_matrix, classification_report
```

```python
def very_confusion_matrix(y_true,y_pred, labels=["False","True"]):
    cm= confusion_matrix(y_true, y_pred)
    pred_labels = ['Predicted' + l for l in labels]
    df=pd.DataFrame(cm,index=labels, columns = pred_labels)
    return df
```

```python
#check the confusion matrix precision and recall
very_confusion_matrix(y_test, y_test_pred, labels=['Stay','Leave'])
```

```python
print(classification_report(y_test,y_test_pred))
```

```python
# Now we need to try Cross validation to see if it still 
#perfroms badly
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
#Check if you still get the same results after 5-k crossfold
def build_logistic_regression_model():
    model = Sequential()
    model.add(Dense(1, input_shape=(20,), activation='sigmoid'))
    model.compile(SGD(lr=0.5), 'binary_crossentropy', metrics=['accuracy'])
    return model

model = KerasClassifier(build_fn = build_logistic_regression_model, epochs=10,
                       verbose=0)
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
cv= KFold(5, shuffle=True)
scores = cross_val_score(model,X,y,cv=cv)

print("The cross validation accuracy is {:0.4f} +- {:0.4f}".format(scores.mean(),scores.std()))
```
